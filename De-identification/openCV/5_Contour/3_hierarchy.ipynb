```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
# contours, hierarchy = cv.findContours(img, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
# findContours(이미지, RetrievalMode, ApproximationMode)
# 다시 원점으로 돌아가서 옵션에 대한 보충 설명

# Retrieval 모드
## RETR_LIST: 객체 바깥쪽과 안쪽 외곽선 모두 검색. 계층 구조 생성 X
## RETR_EXTERNAL: 객체 바깥쪽 외곽선만 검색. 계층 구조 생성 X
## RETR_CCOMP: 모든 외곽선을 검색하고 2단계 계층 구조 구성
## RETR_TREE: 모든 외곽선을 검색하고 전체 계층 구조 구성

img = cv.imread('./hierarchy.bmp', 0)
plt.imshow(img, cmap='gray')
```

```python
def drawCont(src, contours):
    for i in range(len(contours)):
        c = tuple(np.random.choice(range(256), size=3))
        color = (int(c[0]), int(c[1]), int(c[2]))
        cv.drawContours(src, contours, i, color, 3)
        src = cv.cvtColor(src, cv.COLOR_BGR2RGB)
        plt.imshow(src)
```

```python
contours, hierarchy = cv.findContours(img, cv.RETR_LIST, cv.CHAIN_APPROX_NONE)
drawCont(img, contours)
hierarchy
```

```python
contours, hierarchy = cv.findContours(img, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_NONE)
drawCont(img, contours)
hierarchy
```

```python
contours, hierarchy = cv.findContours(img, cv.RETR_CCOMP, cv.CHAIN_APPROX_NONE)
drawCont(img, contours)
hierarchy
```

```python
contours, hierarchy = cv.findContours(img, cv.RETR_TREE, cv.CHAIN_APPROX_NONE)
drawCont(img, contours)
hierarchy
```
