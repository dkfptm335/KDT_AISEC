```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
#   :       
#           

img = cv.imread('../samples/blox.jpg')
plt.imshow(img)
```

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
gray = np.float32(gray)

# img, blocksize, ksize, k
dst = cv.cornerHarris(gray, 4, 3, 0.04)

gray.shape, dst.shape, dst.max()
```

```python
a = dst > 0.01 * dst.max()
np.count_nonzero(a)
```

```python
img[dst > 0.01 * dst.max()] = [0, 0, 255]
plt.imshow(img)
```
