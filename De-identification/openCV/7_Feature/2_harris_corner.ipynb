```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
# 해리스 코너 탐지: 영상의 특정 위치를 기준으로 특정 거리만큼 떨어진
# 픽셀과의 밝기 차이를 계산해서 모든 방향으로 차이값이 크게 나타나면 코너로 인식

img = cv.imread('../samples/blox.jpg')
plt.imshow(img)
```

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
gray = np.float32(gray)

# img, blocksize, ksize, k
dst = cv.cornerHarris(gray, 4, 3, 0.04)

gray.shape, dst.shape, dst.max()
```

```python
a = dst > 0.01 * dst.max()
np.count_nonzero(a)
```

```python
img[dst > 0.01 * dst.max()] = [0, 0, 255]
plt.imshow(img)
```
