```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
# orb = fast + brief
# 실시간 처리에 적합! 빠르고, 방향, 크기, 노이즈에 강함
# 키포인트를 2진수로 구성된 벡터를 사용
```

```python
img = cv.imread('../samples/blox.jpg', cv.IMREAD_GRAYSCALE)

orb = cv.ORB_create()
kp = orb.detect(img, None)

kp, des = orb.compute(img, kp)

img2 = cv.drawKeypoints(img, kp, None, color=(0, 255, 0), flags=0)
plt.imshow(img2), plt.show()
```
