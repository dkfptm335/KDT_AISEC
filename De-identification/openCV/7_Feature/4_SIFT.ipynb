```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('../samples/home.jpg')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
sift = cv.SIFT_create()
kp = sift.detect(img, None)

img2 = cv.drawKeypoints(gray, kp, img, flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(img2)
```

```python
fast = cv.FastFeatureDetector_create(threshold=70)
kp = fast.detect(img, None)
img3 = cv.drawKeypoints(gray, kp, img, flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(img3)
```
