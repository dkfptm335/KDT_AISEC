```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread('../samples/messi5.jpg')

lower1 = cv.pyrDown(img)
lower2 = cv.pyrDown(lower1)

plt.subplot(311), plt.imshow(img)
plt.subplot(312), plt.imshow(lower1)
plt.subplot(313), plt.imshow(lower2)

plt.rcParams["figure.figsize"] = (30, 10)
plt.show()
```
