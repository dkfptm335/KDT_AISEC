```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('../samples/blox.jpg', 0)

plt.imshow(img, cmap='gray')
```

```python
# FAST: 주변 16개의 픽셀 중에서 '점의 밝기+임계값'보다 큰 픽셀이 9개 이상 연속으로 나타나면 그 점을 코너로 판단

fast = cv.FastFeatureDetector_create(threshold=40, nonmaxSuppression=True, type=cv.FastFeatureDetector_TYPE_9_16)
kp = fast.detect(img, None)

img2 = cv.drawKeypoints(img, kp, None, color=(255, 0, 0))

plt.imshow(img2)
```

```python
# NMS: Non-Maximum Suppression, 겹치는 영역 중 가장 좋은 영역 하나를 추린다
fast.setNonmaxSuppression(1)
kp = fast.detect(img, None)
img3 = cv.drawKeypoints(img, kp, None, color=(255, 0, 0))
plt.imshow(img3)
```
