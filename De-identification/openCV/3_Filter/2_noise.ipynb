```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 노이즈 추가 - 가우시안 노이즈

img = cv.imread('../samples//lena.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

fig = plt.figure(figsize=(10, 10))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

mean = 0
var = 500
sigma = var ** 0.5
gaussian = np.random.normal(mean, sigma, (img.shape[0], img.shape[1], img.shape[2]))

noisy_img = np.zeros(img.shape, np.float32)
noisy_img = img + gaussian

ax1.imshow(img)
ax2.imshow(noisy_img)

cv.imwrite('noisy_lena.jpg', noisy_img)
```

```python
# median blurring

img = cv.imread('noisy_lena.jpg', cv.IMREAD_GRAYSCALE)

fig = plt.figure(figsize=(10, 10))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

blur = cv.medianBlur(img, 13)

ax1.imshow(img, cmap='gray')
ax2.imshow(blur, cmap='gray')
```

```python
# gaussian blurring

img = cv.imread('noisy_lena.jpg')

fig = plt.figure(figsize=(10, 10))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

blur = cv.GaussianBlur(img, (13, 13), 0)

ax1.imshow(img, cmap='gray')
ax2.imshow(blur, cmap='gray')

cv.imwrite('gaussian_blur_lena.jpg', blur)
```

```python
# bilateral blurring

img = cv.imread('noisy_lena.jpg', cv.IMREAD_GRAYSCALE)

fig = plt.figure(figsize=(10, 10))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

blur = cv.bilateralFilter(img, 7, 75, 75)

ax1.imshow(img, cmap='gray')
ax2.imshow(blur, cmap='gray')
```
