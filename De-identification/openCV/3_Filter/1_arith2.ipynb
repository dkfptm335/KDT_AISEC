```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(10, 10))
ax1 = plt.subplot(1, 2, 1)
ax2 = plt.subplot(1, 2, 2)

img1 = cv.imread('../samples/lena.jpg')
img1 = cv.cvtColor(img1, cv.COLOR_BGR2RGB)

img2 = np.zeros((512, 512, 3), np.uint8)
cv.rectangle(img2, (162, 162), (350, 350), (255, 255, 255), -1)

ax1.imshow(img1)
ax2.imshow(img2)
```

```python
fig = plt.figure(figsize=(10, 10))
ax1 = plt.subplot(1, 2, 1)
ax2 = plt.subplot(1, 2, 2)

sub1 = cv.subtract(img1, img2)
sub2 = cv.subtract(img2, img1)

ax1.imshow(sub1)
ax2.imshow(sub2)
```

```python
filter = np.zeros((170, 170, 3), np.uint8)
filter += 255

for i in range(0, 3):
    for j in range(0, 3):
        sub1 = img1[i*170:(i+1)*170, j*170:(j+1)*170, :]
        diff = cv.absdiff(sub1, filter)
        img1[i*170:(i+1)*170, j*170:(j+1)*170, :] = diff

plt.imshow(img1)
```
