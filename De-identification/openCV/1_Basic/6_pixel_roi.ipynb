```python
# 
# cv.rectangle(img,  ,  , , )
cv.rectangle(img, (330, 280), (390, 340), (0, 255, 0), 3)

plt.imshow(img)
```

```python
def show_img(img):
    b, g, r = cv.split(img)
    img = cv.merge((r, g, b))
    plt.imshow(img)

show_img(img)
```

```python
imgBGR = cv.imread('../samples/messi5.jpg')
imgRGB = cv.cvtColor(img, cv.COLOR_BGR2RGB)
imgGray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

plt.imshow(imgBGR)
plt.imshow(imgRGB, interpolation='bicubic')
plt.imshow(imgGray, cmap='gray')
```
