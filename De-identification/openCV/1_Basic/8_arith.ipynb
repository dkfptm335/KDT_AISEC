```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

def show_img(img):
    b, g, r = cv.split(img)
    img = cv.merge((r, g, b))
    plt.imshow(img)
```

```python
x = np.uint8([250])
y = np.uint8([10])

# cv.add() 250+10 = 260 → 255
# numpy 더하기 250+10 = 260 → 255가 넘으면 더한 값에 mod 256 연산

cv.add(x,y), x+y
```

```python
img1 = cv.imread('../samples/ml.png')
img2 = cv.imread('../samples/opencv-logo.png')
```

```python
show_img(img1)
```

```python
show_img(img2)
```

```python
img1.shape, img2.shape
```

```python
img2_rs = cv.resize(img2, dsize=(308, 380), interpolation=cv.INTER_AREA)
img2_rs.shape
```

```python
show_img(img2_rs)
```

```python
just_add = cv.add(img1, img2_rs)

show_img(just_add)
```

```python
weighted = cv.addWeighted(img1, 0.7, img2_rs, 0.3, 0)

show_img(weighted)
```
