```python
import pandas as pd

data = pd.read_csv('winequalityN.csv')
```

```python
data
```

```python
data['type'].value_counts()
```

```python
import numpy as np
```

```python
data['type'] = np.where(data['type'] == 'white', 1, 0).astype(int)
```

```python
data.head()
```

```python
data = data.dropna(axis=0)
```

```python
data.isnull().sum()
```

```python
data['quality'].value_counts()
```

```python
x = data.drop(['quality'], axis=1).values
x
```

```python
y = data['quality'].values
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
!pip install lazypredict
```

```python
from lazypredict.Supervised import LazyClassifier
```

```python
clf = LazyClassifier(verbose=0, predictions=True)
```

```python
models, predictions = clf.fit(x_train, x_test, y_train, y_test)
```

```python
models
```

```python
model = RandomForestClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report

for model_name in predictions.columns.tolist():
    print(f'{model_name}')
    print(classification_report(y_test, predictions[model_name]))
```

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.feature_selection import SelectFromModel

fsel = ExtraTreesClassifier().fit(x, y)
model_new = SelectFromModel(fsel, prefit=True)
x_new = model_new.transform(x)
nb_features = x_new.shape[1]
```
