```python
# Data preprocessing
```

```python
import pandas as pd

data = pd.read_csv('winequalityN.csv')
```

```python
data
```

```python
data['type'].value_counts()
```

```python
import numpy as np
```

```python
data['type'] = np.where(data['type'] == 'white', 1, 0).astype(int)
```

```python
data.head()
```

```python
data = data.dropna(axis=0)
```

```python
data.isnull().sum()
```

```python
data['quality'].value_counts()
```

```python
# one-hot encoding
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
# -3을 하는 이유는 원 핫 인코딩을 위해 최솟값을 0으로 만들기 위해서이다.
```

```python
y = to_categorical(data.loc[:, 'quality'] - 3)
```

```python
x = data.drop(['quality'], axis=1).values
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(x)

x_scaled = scaler.fit_transform(x)
x = x_scaled
```

```python
# Split Data

from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
len(x_train), len(x_test)
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
model = Sequential()
```

```python
model.add(Dense(512, activation='relu', input_dim=12))
model.add(Dropout(0.2))
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(16, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(7, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
```

```python
# Learning
```

```python
model.fit(x_train, y_train, epochs=200)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y2 = np.zeros(y_pred.shape, dtype='float32')

for i in range(len(y_pred)):
    j = np.argmax(y_pred[i])
    y2[i][j] = 1

y2
```

```python
accuracy_score(y_test, y2)
```
